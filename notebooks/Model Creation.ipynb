```python
X = dataset[['CPI', 'Consumer Sentiment', 'Durable Goods Orders', 'Treasury Bond Rate', 'Housing Sales', 'Sentiment Deviation from Mean']]
y = dataset['Level']

X.loc[:, 'FOMC_CPI'] = X.loc[:, 'Sentiment Deviation from Mean'] * X.loc[:, 'CPI']
X.loc[:, 'FOMC_Consumer Sentiment'] = X.loc[:, 'Sentiment Deviation from Mean'] * X.loc[:, 'Consumer Sentiment']
X.loc[:, 'FOMC_Durable Goods Orders'] = X.loc[:, 'Sentiment Deviation from Mean'] * X.loc[:, 'Durable Goods Orders']
X.loc[:, 'FOMC_Treasury Bond Rate'] = X.loc[:, 'Sentiment Deviation from Mean'] * X.loc[:, 'Treasury Bond Rate']
X.loc[:, 'FOMC_Housing'] = X.loc[:, 'Sentiment Deviation from Mean'] * X.loc[:, 'Housing Sales']

# Step 4: Create and train the model
X = sm.add_constant(X)  # Add a constant term to the predictor variables
model4 = sm.OLS(y, X)
results4 = model4.fit()

# Step 5: Display model summary
print(results4.summary())
```

```python
# Step 1: Initialize selected_vars with all variables from X
selected_vars = list(X.columns)

# Step 2-5: Perform reverse variable selection
while True:
    # Step 2: Compute p-values for each variable
    p_values = results4.pvalues[1:]  # Exclude the constant term

    # Step 3: Identify the variable with the highest p-value
    max_p_value = p_values.max()
    if max_p_value > 0.05:
        # Step 4: Remove the variable with the highest p-value
        var_to_remove = p_values.idxmax()
        selected_vars.remove(var_to_remove)

        # Step 5: Rebuild the model
        X_selected = X[selected_vars]
        X_selected = sm.add_constant(X_selected)
        model4 = sm.OLS(y, X_selected)
        results4 = model4.fit()
    else:
        break

# Step 6: Display the final model summary
print(results4.summary())
```
